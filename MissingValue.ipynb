```python
import numpy as np
import pandas as pd
from sklearn import datasets
from io import StringIO
from sklearn.model_selection import train_test_split
```

```python
csv_data='''A,B,C,D,E
            5.0,2.0,3.0,,6
            1.0,6.0,,8.0,5
            0.0,11.0,12.0,4.0,5
            3.0,,3.0,5.0,
            5.0,1.0,4.0,2.0,4
            '''
```

```python
df=pd.read_csv(StringIO(csv_data))
df
```

```python
###Pandas的dropna()函數可以刪除欄位###
```

```python
#此函數預設值是只要有任一欄位有空值就會整筆刪掉
df.dropna()
```

```python
#how='all' 要整欄都為空才刪掉
df.dropna(how='all')
```

```python
#當該筆資料的C欄值為空則刪除
df.dropna(subset=['C'])
```

```python
###fillna函式可以補值###
```

```python
#空值補0
df.fillna(0)
```

```python
#B欄位空值用其餘B欄位的值的平均數填入
df['B']=df['B'].fillna(df['B'].mean())
df
```

```python
#C欄位空值用其餘C欄位的值的眾數填入
df['C']=df['C'].fillna(df['C'].mean())
df
```
